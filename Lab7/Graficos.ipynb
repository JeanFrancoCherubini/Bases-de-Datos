```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
tuplasActor100=856421
tuplasActor1k=440234
tuplasActor10k=197219
tuplasActor=[tuplasActor100,tuplasActor1k,tuplasActor10k]

tuplasPersonaje100=2170454
tuplasPersonaje1k=944936
tuplasPersonaje10k=372367
tuplasPersonaje=[tuplasPersonaje100,tuplasPersonaje1k,tuplasPersonaje10k]

tuplasPelicula100=72696
tuplasPelicula1k=22490
tuplasPelicula10k=6401
tuplasPelicula=[tuplasPelicula100,tuplasPelicula1k,tuplasPelicula10k]

```

# P2

## Tiempos de ejecucion

```python
#no indexado
personaje100=413.661
persoaje1k=178.434
personaje10k=69.533
personaje=[personaje100,persoaje1k,personaje10k ]
#indexado
personaje100i= 1.031
persoaje1ki= 0.548
personaje10ki= 0.480
personajei=[personaje100i,persoaje1ki,personaje10ki]


plt.figure(figsize=(6,5))
plt.plot(tuplasPersonaje,personaje,'-o',label='Lab7')
plt.plot(tuplasPersonaje,personajei,'-p',label='Lab7_index')
plt.xticks(np.linspace(372367,2170454,7))
plt.title('Tiempo de Ejecución vs Tamaño de la Tabla')
plt.xlabel('10k                           1K                                                            100 \n Cantidad de Tuplas en la Tabla (Personaje)')
plt.ylabel('Tiempo de Ejecución [ms]')
plt.legend(loc='upper left')
pl.savefig('tiemposEjP2' + '.png')
plt.show()
```

# P3

## Consulta de Rango

```python
#no indexado
pelicula100= 15.565
pelicula1k=5.791
pelicula10k= 1.435
pelicula=[pelicula100,pelicula1k,pelicula10k ]
#indexado
pelicula100i=0.044
pelicula1ki= 0.067
pelicula10ki= 0.054
peliculai=[pelicula100i,pelicula1ki,pelicula10ki]


plt.figure(figsize=(6,5))
plt.plot(tuplasPelicula,pelicula,'-o',label='Lab7')
plt.plot(tuplasPelicula,peliculai,'-p',label='Lab7_index')
plt.xticks(np.linspace(tuplasPelicula[0],tuplasPelicula[2],7))
plt.title('Tiempo de Ejecución de Consulta de Rango \n vs Tamaño de la Tabla')
plt.xlabel('10k                    1K                                                                   100 \n Cantidad de Tuplas en la Tabla (Película)')
plt.ylabel('Tiempo de Ejecución [ms]')
plt.legend(loc='upper left')
pl.savefig('tiemposRango' + '.png')
plt.show()
```

## Consulta Join

```python
tuplasJoin100=tuplasActor100*tuplasPersonaje100
tuplasJoin1k=tuplasActor1k*tuplasPersonaje1k
tuplasJoin10k=tuplasActor10k*tuplasPersonaje10k
tuplasJoin=[tuplasJoin100,tuplasJoin1k,tuplasJoin10k]

#no indexado
join100= 6532.759
join1k=3252.081
join10k= 1836.043
join=[join100,join1k,join10k ]
#indexado
join100i=4763.284
join1ki= 2571.060
join10ki= 1104.358
joini=[join100i,join1ki,join10ki]


plt.figure(figsize=(6,5))
plt.plot(tuplasJoin,join,'-o',label='Lab7')
plt.plot(tuplasJoin,joini,'-p',label='Lab7_index')
plt.xticks(np.linspace(tuplasJoin[0],tuplasJoin[2],7))
plt.title('Tiempo de Ejecución de Consulta con Join vs Tamaño de la Tabla')
plt.xlabel('10k          1K                                                                             100 \n Cantidad de Tuplas en la Tabla (Película*Actor)')
plt.ylabel('Tiempo de Ejecución [ms]')
plt.legend(loc='upper left')
pl.savefig('tiemposJoin' + '.png')
plt.show()
```
